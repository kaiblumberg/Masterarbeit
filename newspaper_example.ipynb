```python
# Import required packages
import pandas as pd
```

```python
df = pd.read_csv('c:\\Friendcloud\\_University\\_Masterarbeit\\Data\\Example Old\\selection-3.csv', sep=';')
df.rename(columns={'year;\"publisher\";\"length\";\"text\"':'rest'}, inplace=True)
df = df.filter(['rest'])
df = df[df.rest.str.contains(';', na=False)]
df = df.dropna()
```

```python
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
```

```python
df
```

```python
df['rest'] = df['rest'].str.replace('"', '')
```

```python
df[['year', 'publisher', 'length', 'text']] = df['rest'].str.split(pat=';', n=3, expand=True)
df.drop('rest', axis=1, inplace=True)
```

```python
# clean publisher data
df['publisher'] = df['publisher'].str.lower()
df['publisher'] = df['publisher'].str.replace('ä', 'ae')
df['publisher'] = df['publisher'].str.replace('ö', 'oe')
df['publisher'] = df['publisher'].str.replace('ü', 'ue')
df['publisher'] = df['publisher'].str.replace('Ä', 'Ae')
df['publisher'] = df['publisher'].str.replace('Ö', 'Oe')
df['publisher'] = df['publisher'].str.replace('Ü', 'Ue')
df['publisher'] = df['publisher'].str.replace('ß', 'ss')
```

```python
df
```

```python
df.to_csv('c:\\Friendcloud\\_University\\_Masterarbeit\\Data\\Source\\selection-3X.csv', sep=';', encoding='utf-8-sig', index=False)
```
