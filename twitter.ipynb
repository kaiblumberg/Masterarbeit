```python
df_category = pd.read_csv('c:\\Friendcloud\\_University\\_Masterarbeit\\Data\\Source\\Overview_Companies_TwitterLinkedIn.csv', sep=';')
```

```python
df_category = df_category[['account','Produktart','category']]
```

```python
df_category
```

```python
df_category['account'] = df_category['account'].str.split('/').str[3]
df_category['account'] = df_category['account'].str.split('?').str[0]
```

```python
df_category
```

```python
df_overview = pd.merge(df_overview, df_category, how='left')
df_overview.sort_values('tweets')
```

```python
df_overview.drop(df_overview[df_overview.account == 'MoolecScience'].index, inplace=True)
```

```python
df_overview.reset_index(inplace=True)
df_overview.drop('index', axis=1, inplace=True)
```

```python
df_overview
```

```python
files = sorted(Path('c:\\Friendcloud\\_University\\_Masterarbeit\\Data\\Source\\Twitter').glob('*.csv'))

df_tweets = pd.DataFrame()
df_list = []
for file in files:
    df_company = pd.read_csv(file)
    file_name = file.name.split('.', 1)[0]
    df_company['account'] = file_name
    df_list.append(df_company)

df_tweets = pd.concat(df_list)
df_tweets.reset_index(inplace=True)
df_tweets.drop('index', axis=1, inplace=True)
```

```python
df_tweets
```

```python
df = pd.merge(df_overview, df_tweets)
```
